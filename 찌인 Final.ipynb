```python
from krwordrank.word import KRWordRank
from konlpy.tag import Okt


def KeyWord(x):
  sentence = []
  wordrank_extractor = KRWordRank(
    min_count = 1, # 단어의 최소 출현 빈도수 (그래프 생성 시)
    max_length = 10, # 단어의 최대 길이
    verbose = True
    )

  beta = 0.85    # PageRank의 decaying factor beta
  max_iter = 20
  keywords, rank, graph = wordrank_extractor.extract(x, beta, max_iter)
  word_list = list()
  for word, r in sorted(keywords.items(), key=lambda x:x[1], reverse=True)[:30]:
    if r >=1:
      word_list.append(word)
  sent = ' '.join(word_list)
  sentence.append(sent)
  okt = Okt()
  OKT = okt.pos(sent)
  keyword_list = []
  # 명사 추출
  for word, tag in OKT:
    if (tag not in ['Josa']) and (tag not in ['Suffix']):
      keyword_list.append(word)
  return keyword_list
```

```python
df['KeyWord_2'] = df['KeyWord_1'].apply(KeyWord)
```

```python
import jpype
from konlpy.tag import Okt
okt = Okt()

df['상품명_전처리'] = df['상품명'].map(okt.nouns)
df['상품명_전처리']
```

```python
tfidf_vectorizer_2 = TfidfVectorizer()

tfidf_matrix_2 = tfidf_vectorizer_2.fit_transform(df['상품명_전처리'].map(' '.join))
tfidf_array_2 = tfidf_matrix_2.toarray()
tfidf_df_2 = pd.DataFrame(tfidf_array_2, columns=tfidf_vectorizer_2.get_feature_names_out())
tfidf_df_2
```

```python
from sklearn.metrics.pairwise import cosine_similarity

cosine_sim = cosine_similarity(tfidf_array) # 책소개 키워드
cosine_sim_1 = cosine_similarity(tfidf_array_1) # 키워드
cosine_sim_2 = cosine_similarity(tfidf_array_2) # 상품명
```

```python
pd.DataFrame(cosine_sim,index = df['상품명'],columns=df['상품명'])
```

```python
pd.DataFrame(cosine_sim_1,index = df['상품명'],columns=df['상품명'])
```

```python
pd.DataFrame(cosine_sim_2,index = df['상품명'],columns=df['상품명'])
```

```python
# cosine_sim -> 책소개
# cosine_sim_1 -> 키워드Pick
# cosine_sim_2 -> 상품명
sim = 0.4*cosine_sim + 0.5*cosine_sim_1 + 0.1*cosine_sim_2
sim
```

```python
query_index = -1
top_k = 10

# 쿼리 인덱스에 대한 유사도 값을 가져옴
query_similarity = sim[query_index]

# 코사인 유사도 값을 기준으로 유사도가 높은 순서대로 정렬된 인덱스를 얻음
sorted_indices = np.argsort(query_similarity)[::-1]

print('검색한 콘텐츠')
print(df['상품명'].iloc[query_index])
print()
print('출력한 콘텐츠')
df['상품명'].iloc[sorted_indices[1:top_k +1]]
```
