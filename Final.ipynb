## TF-IDF

```python
type(list(df['책소개 키워드'][0]))
```

```python
def to_list(x):
  return list(x)
df['책소개 키워드']=df['책소개 키워드'].apply(to_list)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer()
```

```python
tfidf_matrix = tfidf_vectorizer.fit_transform(df['책소개 키워드'].map(' '.join))
```

```python
tfidf_array = tfidf_matrix.toarray()
tfidf_df = pd.DataFrame(tfidf_array, columns=tfidf_vectorizer.get_feature_names_out())
```

```python
tfidf_df
```

```python
tfidf_vectorizer_1 = TfidfVectorizer()

tfidf_matrix_1 = tfidf_vectorizer_1.fit_transform(df['키워드'].map(' '.join))
tfidf_array_1 = tfidf_matrix_1.toarray()
tfidf_df_1 = pd.DataFrame(tfidf_array_1, columns=tfidf_vectorizer_1.get_feature_names_out())
```

```python
tfidf_df_1
```

```python
import jpype
from konlpy.tag import Okt
okt = Okt()

df['상품명_전처리'] = df['상품명'].map(okt.nouns)
df['상품명_전처리']
```

```python
tfidf_vectorizer_2 = TfidfVectorizer()

tfidf_matrix_2 = tfidf_vectorizer_2.fit_transform(df['상품명_전처리'].map(' '.join))
tfidf_array_2 = tfidf_matrix_2.toarray()
tfidf_df_2 = pd.DataFrame(tfidf_array_2, columns=tfidf_vectorizer_2.get_feature_names_out())
tfidf_df_2
```

```python
from sklearn.metrics.pairwise import cosine_similarity

cosine_sim = cosine_similarity(tfidf_array) # 책소개 키워드
cosine_sim_1 = cosine_similarity(tfidf_array_1) # 키워드
cosine_sim_2 = cosine_similarity(tfidf_array_2) # 상품명
```

```python
pd.DataFrame(cosine_sim)
```

```python
pd.DataFrame(cosine_sim_1)
```

```python
pd.DataFrame(cosine_sim_2)
```

```python
sim = 0.25*cosine_sim + 0.5*cosine_sim_1 + 0.25*cosine_sim_2
sim
```

```python
query_index = 1
top_k = 10

# 쿼리 인덱스에 대한 유사도 값을 가져옴
query_similarity = sim[query_index]

# 코사인 유사도 값을 기준으로 유사도가 높은 순서대로 정렬된 인덱스를 얻음
sorted_indices = np.argsort(query_similarity)[::-1]

print('검색한 콘텐츠')
print(df['상품명'].iloc[query_index])
print()
print('출력한 콘텐츠')
df['상품명'].iloc[sorted_indices[1:top_k +1]]
```
