```python
베스트셀러 = pd.read_excel(path + '/데이터 전처리/책소개 키워드 추출.xlsx')
```

```python
best = 베스트셀러.iloc[:, 1:]
best
```

```python
best = best.drop_duplicates(keep='last',ignore_index=True)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
count_vect_keyword = CountVectorizer()
book_mat_keyword = count_vect_keyword.fit_transform(best['키워드'])                             
```

```python
book_mat_keyword.toarray()
```

```python
count_vect_keyword.vocabulary_
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_sim_keyword = cosine_similarity(book_mat_keyword, book_mat_keyword)
print(cosine_sim_keyword.shape)
print(cosine_sim_keyword[:10])
```

```python
cosine_sim_keyword_sroted_index = cosine_sim_keyword.argsort()[:,::-1]
print(cosine_sim_keyword_sroted_index[:5])
```

```python
title = '브랜드 설계자'
```

```python
book_title = best[best['상품명']==title]
book_index = book_title.index.values
similar_indexs = cosine_sim_sroted_index[book_index, :10]
print(similar_indexs)
```

```python
def find_sim_book_keyword(best, sorted_index, title, top_n):
    book_title = best[best['상품명']==title]
    book_index = book_title.index.values
    similar_indexs = sorted_index[book_index, :top_n]
    print(similar_indexs)
    return best.iloc[similar_indexs[0], :]
```

```python
find_sim_book_keyword(best, cosine_sim_keyword_sroted_index, title, 10)
```
